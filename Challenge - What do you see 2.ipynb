```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
plt.rcParams['figure.figsize'] = (10.0, 7.0)
plt.rcParams.update({'font.size': 15})
df = pd.read_csv('cat.csv')
df.head(10)
```

```python
df.columns
```

```python
plt.hist(df['distance'],bins = 10)
plt.xlabel('Distance')
plt.xlim(0,80)
plt.ylabel('Frequency')
plt.title('Distance Frequency Distribution');
```

```python
df.country_name.value_counts().head(50).plot(kind = "bar")
plt.xlabel('Landslides')
plt.ylabel('Count')
plt.title('Landslide by Countries Count');
```

```python
df.country_name.value_counts().head(50)[1:50].plot(kind = "bar")
plt.xlabel('Landsides')
plt.ylabel('Count')
plt.title('Landslides by Countries Count');
```

```python
df.groupby('country_name').fatalities.mean().plot(kind = "bar")
plt.xlabel('Country')
plt.ylabel('Average Fatalities')
plt.title('Average Fatalities by Country');
```

```python
df.boxplot(column = 'distance', by = 'country_name')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Distance');
```

```python
df.groupby('date').distance.mean().plot(kind = "line")
plt.xlabel('Date')
plt.ylabel('Distance')
plt.title('Distance Over Time')
tick_val = [0, 25, 50, 75]
tick_lab = ['0','25', '50', '75'] 
plt.yticks(tick_val, tick_lab);
```

```python
df.groupby('date').fatalities.mean().plot(kind = "line")
plt.xlabel('Date')
plt.ylabel('Fatalities')
plt.title('Fatalities Over Time')
tick_val = [0, 50, 100, 300]
plt.yticks(tick_val);
```
