```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
df = pd.read_csv('data.csv')
df
```

```python
plt.scatter(
    x=df['INTAGE'],
    y=df['HRS1'],
    color='purple',
)
plt.ylabel("Hours worked the previous week")
plt.xlabel('Age in Years')
plt.title('Hours Worked Per Respondant\'s Age')
plt.show()
```

The majority of the population apears to work around 20-60 hours during the week through the ages of 40 - 70. 

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.plot(df['AGE'],color='purple')
plt.ylabel('Age of Respondants')
plt.title('Age in Years')

plt.subplot (1,2,2)
plt.scatter(x=df['AGE'], y = df['HOMPOP'], color='green')
plt.ylim([0, 15])
plt.ylabel('Household Members')
plt.xlabel ('Age')
plt.title('Household Members')
plt.tight_layout()
plt.show()
```

The first line plot here showcasing ages over the number of respondants is not as textbook as the examples in the course due to the vast number of responses and age varations. The second scatter subplot however, shows us a more detailed outline of the number of household members by age1- group. It appears the majority range for all age goups is 1-2 household members. 

```python
plt.hist(df['DEGREE'], bins= 5, color='blue')
plt.xlim([0,10])
plt.title('Respondant\'s Highest Degree Earned')
plt.xlabel('Degree Groupings')
plt.show()
```

A majority of the respondent's completed high school, while the second largest grouping is those who had completed their bachelor's degree. 

```python
plt.hist(df['DEGREE'], density = True, color='blue', bins=np.arange(0, 10), alpha=.5) 
plt.hist(df['MADEG'], density = True, color='yellow', bins=np.arange(0, 10), alpha=.5)
plt.hist(df['PADEG'], density = True, color='orange', bins=np.arange(0, 10), alpha=.5)
plt.title('Respondents Highest Degree Earned vs. Parent\'s Highest Degree Earned')
plt.xlabel('Degree Groupings')

plt.show()
```

The respondants overall have completed more schooling than that of both their mother's and father's.  It also appears, that respondants mother's had completed further schooling than respondant's father's. However, more respondents did not know, or found it innaplicable to discuss their father's highest degree earned.
